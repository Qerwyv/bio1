```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv', sep=',')
pd.set_option('display.expand_frame_repr', False)
print(df)
```

```python
from matplotlib import style
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
style.use('ggplot')
x = list(df.columns.values)[1:]
months = 'Вересень Жовтень Листопад Грудень Січень Лютий Березень'.split(' ')
months.reverse()
colors = 'g r c m y k b'.split(' ')
for month in range(0, 7):
    y = df.iloc[month, 1:]
    plt.plot(x,y,color=colors.pop(),label=months.pop(), linewidth=1)
plt.title('Середньомісячні концентрації забруднювальних речовин\nв атмосферному повітрі м. Вінниця за вересень-березень 2017-2018 року')
plt.ylabel('мг/м3', fontsize=13)
plt.xlabel('Домішки', fontsize=17)
plt.legend()
plt.grid(True,color='k')
plt.tight_layout()
plt.show()
```
